# Part 1

## Step 1: Create a Matrix with Student Grades

```python
import numpy as np
student_names= []
grades_list= []
for i in range (10):
    name = input(f"Enter name for student {i+1}: ")
    student_names.append(name)
    print(f"  Enter grades for {name}:")
    homework = float(input("    Homework: "))
    project = float(input("    Project: "))
    exam1 = float(input("    Exam 1: "))
    exam2 = float(input("    Exam 2: "))
    final_exam = float(input("    Final Exam: "))
    grades_list.append([homework, project, exam1, exam2, final_exam])
```

## Step 2: Create weights 

```python
weights = [0.15, 0.15, 0.20, 0.20, 0.30]
```

## Step 3: Calculate final grade for each student 

```python
final_grades= []
for student_grades in grades_list:
    final = (student_grades[0] * weights[0] + 
             student_grades[1] * weights[1] + 
             student_grades[2] * weights[2] + 
             student_grades[3] * weights[3] + 
             student_grades[4] * weights[4])
    final_grades.append(final)
```

## Step 4: Convert to letter grades

```python
letter_grades= []
for grade in final_grades:
    if grade >= 90:
        letter_grades.append('A')
    elif grade >= 80:
        letter_grades.append('B')
    elif grade >= 70:
        letter_grades.append('C')
    elif grade >= 60:
        letter_grades.append('D')
    else:
        letter_grades.append('F')
```

## Step 5: Print students names and grades 

```python
print("\n==== Final Grades ====")
for i in range(10):
    print(f"{student_names[i]}: {final_grades[i]:.1f} ({letter_grades[i]})")
```
